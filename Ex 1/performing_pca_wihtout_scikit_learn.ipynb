#Common PCA Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_normalizing(data):
  meanValue = np.mean(data,axis=0)
  nor_data = data - meanValue
  return nor_data,meanValue
```

```python
def cal_cov_matrix(data):
    num_samples = data.shape[0]
    cov_matrix = np.dot(data.T,data)/(num_samples-1)
    return cov_matrix
```

```python
def perfomrPCA(data,prinicpal_Components):
  normalizedData, meanValue = mean_normalizing(data)
  cov_matrix = cal_cov_matrix(normalizedData)
  eigenValue, eigenVector = np.linalg.eig(cov_matrix)
  sorted_indices = np.argsort(eigenValue)[::-1]
  eigenValue = eigenValue[sorted_indices]
  eigenVector = eigenVector[:,sorted_indices]
  selected_eigenv = eigenVector[:,:prinicpal_Components]
  pca_res = np.dot(normalizedData,selected_eigenv)
  return pca_res,eigenValue,selected_eigenv,meanValue
```

```python
def explainedVariance(eigenValue):
  totalSum = np.sum(eigenValue)
  explained_variance_ratio = eigenValue/totalSum
  cumulative_explained_variance = np.cumsum(explained_variance_ratio)
  return cumulative_explained_variance
```

#Plotting Functions

```python
def explainedVariancePlot(eigen_Value):
  cumExplainedVariance = explainedVariance(eigen_Value)
  plt.plot(range(1, len(cumExplainedVariance) + 1), cumExplainedVariance, marker='o')
  plt.title('Cumulative Explained Variance')
  plt.xlabel('Number of Principal Components')
  plt.ylabel('Cumulative Explained Variance')
  plt.show()
```

```python
def scatterPlot(data):

  plt.scatter(data[:,0], data[:, 1])
  plt.xlabel("Feature 1")
  plt.ylabel("Feature 2")
  plt.title("DATA")
```

```python
def Heatmap(X,eig_vecs):
  # ---- Standardize data (important for PCA) ----
  X_meaned = X - np.mean(X, axis=0)

  # ---- Select top k eigenvectors (say k=2 or k=3) ----
  k = 2
  eig_vecs_k = eig_vecs[:, :k]

  # ---- Project data to PCA space ----
  X_pca = np.dot(X_meaned, eig_vecs_k)

  # ---- Plot heatmaps ----
  plt.figure(figsize=(12,5))

  # Original Data Heatmap
  plt.subplot(1,2,1)
  plt.imshow(X, aspect='auto')
  plt.colorbar()
  plt.title("Original Data Heatmap")
  plt.xlabel("Features")
  plt.ylabel("Samples")

  # PCA Data Heatmap
  plt.subplot(1,2,2)
  plt.imshow(X_pca, aspect='auto')
  plt.colorbar()
  plt.title("PCA Transformed Data Heatmap")
  plt.xlabel("Principal Components")
  plt.ylabel("Samples")

  plt.tight_layout()
  plt.show()
```

#First Classwork Problem

```python
PrincipalComponents = 1
data = np.array([
    [4,11],
    [8,4],
    [13,5],
    [7,14]
])
```

```python
pcaResult, eigen_Value, chosen_eigneVector, mean = perfomrPCA(data, PrincipalComponents)
```

```python
print("Original shape" , data.shape)
print("PCA Shape", pcaResult.shape)
print("All eigen values: ", eigen_Value)
print("Eigen Vectors: ", chosen_eigneVector)
print("Mean: ", mean)
```

```python
print("ORiginal data\nX Y\n")
for i in data:
  print(i)
print('\n')
print("PCA DATA\n")
for i in pcaResult:
  print(i)
```

```python
explainedVariancePlot(eigen_Value)
```

```python
scatterPlot(data)
```

```python
Heatmap(data, chosen_eigneVector)
```
