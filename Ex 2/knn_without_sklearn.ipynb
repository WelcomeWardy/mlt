```python
import numpy as np
```

```python
# Feature matrix (X) — each row is a data point
X_train = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [6, 5],
    [7, 7],
    [8, 6]
])

# Binary class labels (y)
# 0 -> Class A, 1 -> Class B
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def knn_predict(X_train, y_train, x_test, k):
    distances = []

    # Step 1: Compute distance from test point to all training points
    for i in range(len(X_train)):
        dist = euclidean_distance(X_train[i], x_test)
        distances.append((dist, y_train[i]))

    # Step 2: Sort distances
    distances.sort(key=lambda x: x[0])

    # Step 3: Select k nearest neighbors
    k_nearest = distances[:k]

    # Step 4: Majority voting
    class_votes = [label for (_, label) in k_nearest]
    predicted_class = max(set(class_votes), key=class_votes.count)

    return predicted_class, k_nearest
```

```python
x_new = np.array([5, 5])
k = 3
```

```python
predicted_class, neighbors = knn_predict(X_train, y_train, x_new, k)
```

```python
print("New data point:", x_new)
print("K nearest neighbors (distance, class):")

for d, c in neighbors:
    print(f"Distance: {d:.2f}, Class: {c}")

print("\nPredicted Class:", predicted_class)
```
