```python
import numpy as np

# 1. Training data (x: input, y: output)
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# 2. Initialize parameters
m = 0      # slope
b = 0      # intercept
alpha = 0.01   # learning rate
epochs = 1000
n = len(x)

# 3. Gradient Descent algorithm
for i in range(epochs):
    y_pred = m * x + b

    # Compute gradients
    dm = (-2/n) * np.sum(x * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Update parameters
    m = m - alpha * dm
    b = b - alpha * db

# 4. Final parameters
print("Slope (m):", m)
print("Intercept (b):", b)

# 5. Prediction
x_test = 6
y_test_pred = m * x_test + b
print("Prediction for x = 6:", y_test_pred)
```
